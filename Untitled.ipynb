```python
scaler = StandardScaler()
scaler.fit_transform(x)
```

```python
dft = pd.DataFrame(scaler.fit_transform(x))
```

```python
#PCA
pca = PCA(n_components=20)
pca.fit_transform(dft)
```

```python
pd.DataFrame(pca.fit_transform(x))
```

```python
variance = pca.explained_variance_ratio_
```

```python
plt.bar(np.arange(len(variance)), variance, alpha=0.5, align='center')
plt.xticks(np.arange(len(variance)), ['PC{}'.format(i+1) for i in range(len(variance))])
plt.ylabel('Variance Explained')
plt.xlabel('Principal Component')
plt.title('PCA Bar Plot')
plt.show()
```
